```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('results_4x4.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.bar(data["algorithm"], data["seconds"])

ax.set_title("Alltoall vs Locality-Aware Alltoall (CARC Wheeler)\n4 Nodes, 4 Processes per Node\n")
ax.set_xlabel("Algorithm")
ax.set_ylabel("Seconds (average of 1000 runs)")
plt.tight_layout()
plt.savefig("alltoall_plot.jpg")
plt.show()
```

```python
data2 = pd.read_csv('results_4x4_2.csv')

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.bar(data["algorithm"], data["seconds"], label="original")
ax.bar(data2["algorithm"], data2["seconds"], label="factored out malloc")

ax.set_title("Alltoall vs Locality-Aware Alltoall (CARC Wheeler)\n4 Nodes, 4 Processes per Node\n")
ax.set_xlabel("Algorithm")
ax.set_ylabel("Seconds (average of 1000 runs)")
plt.legend()
plt.tight_layout()
plt.savefig("alltoall_plot2.jpg")
plt.show()
```

```python
data3 = pd.read_csv('results_4x4_3.csv')

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.bar(data["algorithm"], data["seconds"], label="original")
ax.bar(data3["algorithm"], data3["seconds"], label="square and factored out malloc")
ax.bar(data2["algorithm"], data2["seconds"], label="factored out malloc")

ax.set_title("Alltoall vs Locality-Aware Alltoall (CARC Wheeler)\n4 Nodes, 4 Processes per Node\n")
ax.set_xlabel("Algorithm")
ax.set_ylabel("Seconds (average of 1000 runs)")
plt.legend()
plt.tight_layout()
plt.savefig("alltoall_plot3.jpg")
plt.show()
```
