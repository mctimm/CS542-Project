```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_4x4 = pd.read_csv('results_4x4.csv')
data_8x8 = pd.read_csv('results_8x8.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.bar(data_4x4["algorithm"], data_4x4["seconds"])

ax.set_title("Alltoall vs Locality-Aware Alltoall (CARC Wheeler)\n4 Nodes, 4 Processes per Node\n")
ax.set_xlabel("Algorithm")
ax.set_ylabel("Seconds (average of 1000 runs)")
plt.tight_layout()
plt.savefig("alltoall_plot_4x4.jpg")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.bar(data_8x8["algorithm"], data_8x8["seconds"])

ax.set_title("Alltoall vs Locality-Aware Alltoall (CARC Wheeler)\n8 Nodes, 8 Processes per Node\n")
ax.set_xlabel("Algorithm")
ax.set_ylabel("Seconds (average of 1000 runs)")
plt.tight_layout()
plt.savefig("alltoall_plot_8x8.jpg")
plt.show()
```
