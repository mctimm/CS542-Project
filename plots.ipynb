```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_4x4 = pd.read_csv('results_4x4.csv')
data_local = data_4x4[data_4x4["algorithm"] == "AlltoallVarSize"]
data_openmpi = data_4x4[data_4x4["algorithm"] == "MPI_Alltoall"]
data_comm = data_4x4[data_4x4["algorithm"] == "AlltoallVarSize_comm_only"]
data_rsm = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall"]
data_buff = data_4x4[data_4x4["algorithm"] == "AlltoallNoShiftBuffered"]
data_manpack = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall_manpack"]
data_mpipack = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall_mpipack"]
data_manpack_sr = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall_manpack_sendrecv"]
data_sr = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall_sendrecv"]
data_wait = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall_wait"]
data_local_bruck = data_4x4[data_4x4["algorithm"] == "alltoall_local_bruck"]
data_bruck = data_4x4[data_4x4["algorithm"] == "alltoall_bruck"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# ax.plot(data_openmpi["num_doubles_per_proc"]*8/16, data_openmpi["seconds"], 'o-', label='MPI_Alltoall', markersize=5)
# ax.plot(data_local["num_doubles_per_proc"]*8/16, data_local["seconds"], 'o-', label='AlltoallVarSize', markersize=5)
# ax.plot(data_comm["num_doubles_per_proc"]*8/16, data_comm["seconds"], 'o-', label='AlltoallVarSize_comm', markersize=5)
# ax.plot(data_buff["num_doubles_per_proc"]*8/16, data_buff["seconds"], 'o-', label='AlltoallNoShiftBuffered', markersize=5)

ax.plot(data_rsm["num_doubles_per_proc"]*8/16, data_rsm["seconds"], 'o-', label='RSM_Alltoall', markersize=5)
# ax.plot(data_manpack["num_doubles_per_proc"]*8/16, data_manpack["seconds"], 'o-', label='RSM_Alltoall_manpack', markersize=5)
# ax.plot(data_mpipack["num_doubles_per_proc"]*8/16, data_mpipack["seconds"], 'o-', label='RSM_Alltoall_mpipack', markersize=5)
# ax.plot(data_manpack_sr["num_doubles_per_proc"]*8/16, data_manpack_sr["seconds"], 'o-', label='RSM_Alltoall_manpack_sendrecieve', markersize=5)
# ax.plot(data_sr["num_doubles_per_proc"]*8/16, data_sr["seconds"], 'o-', label='RSM_Alltoall_sendrecv', markersize=5)
# ax.plot(data_wait["num_doubles_per_proc"]*8/16, data_wait["seconds"], 'o-', label='RSM_Alltoall_wait', markersize=5)

ax.plot(data_local_bruck["num_doubles_per_proc"]*8/16, data_local_bruck["seconds"], 'o-', label='alltoall_local_bruck', markersize=5)
ax.plot(data_bruck["num_doubles_per_proc"]*8/16, data_bruck["seconds"], 'o-', label='alltoall_bruck', markersize=5)

plt.xlim([2**2, 2**19])
plt.ylim([2**-17, 2**-1])
ax.set_xscale('log', base=2)
ax.set_yscale('log', base=2)
ax.set_title("Alltoall measurements, 4x4")
ax.set_xlabel("Number of bytes in message")
ax.set_ylabel("Seconds (avg. of 1000 trials)")
ax.legend()
plt.savefig("alltoall_4x4.jpg")
plt.show()
```

```python
data_4x4_2 = pd.read_csv('results_4x4_2.csv')
data_openmpi = data_4x4_2[data_4x4_2["algorithm"] == "MPI_Alltoall"]
data_rsm = data_4x4_2[data_4x4_2["algorithm"] == "RSM_Alltoall"]
data_buff = data_4x4_2[data_4x4_2["algorithm"] == "AlltoallNoShiftBuffered"]
data_rsm_comm = data_4x4_2[data_4x4_2["algorithm"] == "RSM_Alltoall_comm"]
data_buff_comm = data_4x4_2[data_4x4_2["algorithm"] == "AlltoallNoShiftBuffered_comm"]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(data_openmpi["num_doubles_per_proc"]*8/16, data_openmpi["seconds"], 'o-', label='MPI_Alltoall', markersize=5)
ax.plot(data_buff["num_doubles_per_proc"]*8/16, data_buff["seconds"], 'o-', label='AlltoallNoShiftBuffered', markersize=5)
ax.plot(data_buff_comm["num_doubles_per_proc"]*8/16, data_buff_comm["seconds"], 'o-', label='AlltoallNoShiftBuffered_comm', markersize=5)
ax.plot(data_rsm["num_doubles_per_proc"]*8/16, data_rsm["seconds"], 'o-', label='RSM_Alltoall', markersize=5)
ax.plot(data_rsm_comm["num_doubles_per_proc"]*8/16, data_rsm_comm["seconds"], 'o-', label='RSM_Alltoall_comm', markersize=5)

plt.xlim([2**2, 2**19])
plt.ylim([2**-17, 2**-1])
ax.set_xscale('log', base=2)
ax.set_yscale('log', base=2)
ax.set_title("Alltoall measurements, 4x4")
ax.set_xlabel("Number of bytes in message")
ax.set_ylabel("Seconds (avg. of 1000 trials)")
ax.legend()
plt.savefig("alltoall_4x4_2.jpg")
plt.show()
```
