```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_4x4 = pd.read_csv('results_4x4.csv')
data_local = data_4x4[data_4x4["algorithm"] == "AlltoallVarSize"]
data_openmpi = data_4x4[data_4x4["algorithm"] == "MPI_Alltoall"]
data_comm = data_4x4[data_4x4["algorithm"] == "AlltoallVarSize_comm_only"]
data_rsm = data_4x4[data_4x4["algorithm"] == "RSM_Alltoall"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(data_local["num_doubles_per_proc"]*8/16, data_local["seconds"], 'bo-', label='AlltoallVarSize', markersize=5)
ax.plot(data_openmpi["num_doubles_per_proc"]*8/16, data_openmpi["seconds"], 'ro-', label='MPI_Alltoall', markersize=5)
ax.plot(data_comm["num_doubles_per_proc"]*8/16, data_comm["seconds"], 'go-', label='AlltoallVarSize_comm', markersize=5)
ax.plot(data_rsm["num_doubles_per_proc"]*8/16, data_rsm["seconds"], 'bo-', label='RSM_Alltoall', markersize=5)

ax.set_xscale('log', base=2)
ax.set_yscale('log', base=2)
ax.set_title("Alltoall measurements, 4x4")
ax.set_xlabel("Number of bytes in message")
ax.set_ylabel("Seconds (avg. of 1000 trials)")
ax.legend()
plt.savefig("alltoall_4x4.jpg")
plt.show()
```
