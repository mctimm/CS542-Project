```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('results_4x4.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.bar(data["algorithm"], data["seconds"])

ax.set_title("Alltoall vs Locality-Aware Alltoall (CARC Wheeler)\n4 Nodes, 4 Processes per Node\n")
ax.set_xlabel("Algorithm")
ax.set_ylabel("Seconds (average of 1000 runs)")
plt.tight_layout()
plt.savefig("alltoall_plot.jpg")
plt.show()
```
